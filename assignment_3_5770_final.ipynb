```python
# Check the updated column before mean imputation
print("before mean imputation null count : ",asthma_selected_df['ADHECR42'].isnull().sum())

# ADHECR42 mean
mean_adhecr42 = asthma_selected_df['ADHECR42'].mean()

# Fill nulls with mean value
asthma_selected_df['ADHECR42'].fillna(mean_adhecr42, inplace=True)

# Check the updated column after mean imputation
print("after mean imputation null count : ",asthma_selected_df['ADHECR42'].isnull().sum())
```

```python
# Detect outliers
def detect_outliers(asthma_data):
    # Calculate Q1 (25th percentile), Q3 (75th percentile), and IQR for each column
    Q1 = asthma_data.quantile(0.25)
    Q3 = asthma_data.quantile(0.75)
    IQR = Q3 - Q1

    # Calculate lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Return boolean DataFrame where True indicates an outlier
    return ((asthma_data < lower_bound) | (asthma_data > upper_bound))

# Select numerical columns only
num_cols = asthma_selected_df.drop(['DUPERSID','SEX', 'ASPRIN53', 'RACETHX'],axis=1)

# Detect outliers in the numeric columns
outliers = detect_outliers(num_cols)

# Count the number of outliers for each row
outlier_count = outliers.sum(axis=1)

# Remove rows where more than 3 attributes have outlier values
asthma_selected_cleaned_df = asthma_selected_df[outlier_count <= 3]

# Number of rows left after removing outliers
remaining_rows = asthma_selected_cleaned_df.shape[0]
print(f"Number of tuples left after removing outliers: {remaining_rows}")
```

```python
# BMINDX col creation
asthma_selected_cleaned_df['BMINDX'] = np.where(asthma_selected_cleaned_df['AGE16X'] > 17, asthma_selected_cleaned_df['BMINDX53'], asthma_selected_cleaned_df['CHBMIX42'])
print(asthma_selected_cleaned_df[['AGE16X','BMINDX53', 'CHBMIX42', 'BMINDX']])
```

```python
# Step 1: Calculate the correlation matrix for numeric attributes
correlation_matrix = asthma_selected_cleaned_df.drop(['DUPERSID','SEX', 'ASPRIN53', 'RACETHX'],axis=1).corr()
print(correlation_matrix)

# Step 2: Identify highly correlated pairs (above 0.8 or below -0.8)
# We're setting correlations of 1.0 (self-correlation) to NaN to exclude them
correlation_matrix_filtered = correlation_matrix.where((correlation_matrix != 1) & ((correlation_matrix > 0.8) | (correlation_matrix < -0.8)))

# Step 3: Display the filtered correlation matrix showing only high correlations
print("Highly correlated pairs (correlation > 0.8 or < -0.8):")
print(correlation_matrix_filtered.dropna(how='all'))  # Drop rows where all values are NaN

# Optional: Display attribute pairs with their correlation values as a list
high_corr_pairs = correlation_matrix_filtered.stack()  # Stacks the non-null correlations into pairs
print("\nAttribute pairs with strong correlations:")
print(high_corr_pairs)
```

```python
# Dropping BMINDX53 and CHBMIX42 since BMINDX now combines both and WAGE16X.
asthma_selected_cleaned_df = asthma_selected_cleaned_df.drop(columns=['CHBMIX42', 'BMINDX53', 'WAGEP16X'])

# Check the remaining columns
asthma_selected_cleaned_df.head()
```

```python
from scipy.stats import chi2_contingency

# Define the pairs of columns to test
pairs = [('SEX', 'RACETHX'), ('SEX', 'ASPRIN53'), ('RACETHX', 'ASPRIN53')]

# Loop through each pair and perform the Chi-square test
for var1, var2 in pairs:
    contingency_table = pd.crosstab(asthma_selected_cleaned_df[var1], asthma_selected_cleaned_df[var2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f"Chi-square Test between {var1} and {var2}")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:")
    print(expected)
    print("\n" + "-"*40 + "\n")
```

Categorical Variable: The p-value between RACETHX and ASPRIN53 is 0.023 which indicates significant correlation. Considering, RACETHX gives more information on understanding asthma prevalence and severity retaining that and dropping ASPRIN53.

```python
# Dropping ASPRIN53
asthma_selected_cleaned_df = asthma_selected_cleaned_df.drop(columns=['ASPRIN53'])

# Check the remaining columns
asthma_selected_cleaned_df.head()
```

```python
# Exclude 'DUPERSID' and other non-numeric categorical attributes
columns_to_exclude = ['DUPERSID', 'SEX', 'RACETHX']  # Add any other categorical variables to this list

# Select numerical columns that are not in the exclusion list
numerical_columns = asthma_selected_cleaned_df.select_dtypes(include=['float64', 'int64']).columns.difference(columns_to_exclude)

# Apply Min-Max scaling to the selected numerical columns
asthma_selected_cleaned_df[numerical_columns] = (asthma_selected_cleaned_df[numerical_columns] - asthma_selected_cleaned_df[numerical_columns].min()) / (asthma_selected_cleaned_df[numerical_columns].max() - asthma_selected_cleaned_df[numerical_columns].min())

# Check the normalized data
print(asthma_selected_cleaned_df[numerical_columns].head())
```
