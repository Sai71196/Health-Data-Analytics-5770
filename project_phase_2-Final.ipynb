FINAL VARIABLES

1. SEX: Hypertension prevalence, risk factors, and outcomes differ between males and females due to biological and hormonal factors.
2. RACETHX: Hypertension is more prevalent in certain racial/ethnic groups (e.g., African Americans) and influences outcomes differently.
3. HIBPDX : The primary variable of interest, representing whether a person has been diagnosed with hypertension.
4. HIBPAGED: Early diagnosis age can indicate a higher risk of severe outcomes or long-term hypertension.
5. BPMLDX : Ensures that blood pressure diagnosis aligns with hypertension diagnosis and reflects clinical evaluation.
6. CHDDX: Hypertension is a major risk factor for coronary heart disease, so this variable provides insight into complications.
7. CHOLDX: High cholesterol often coexists with hypertension and contributes to cardiovascular disease risk.
8. CHOLAGED: Early diagnosis of cholesterol problems may compound hypertension risks.
9. STRKDX: Hypertension is the leading cause of stroke, so this variable reflects complications and severity.
10. EMPST31: Employment status may correlate with stress levels, healthcare access, or socioeconomic factors affecting hypertension.

11. MNHLTH31: Stress and poor mental health are known contributors to hypertension onset and progression.
12. ACTLIM31: Indicates how hypertension or related conditions affect daily living.
13. DIABDX_M18: Hypertension and diabetes often coexist, and the conditions exacerbate each other’s risks.
14. MIDX: Hypertension is a major risk factor for myocardial infarction (heart attack), so this variable reflects severity.
15. OFTSMK53: Smoking is a modifiable risk factor for hypertension and cardiovascular diseases.
16. FAMINC19:Socioeconomic status impacts access to healthcare, medications, and lifestyle choices affecting hypertension.
17. INSCOV19: Insurance coverage reflects access to healthcare services for managing hypertension.
18. RXNUM: Tracks medication burden and treatment adherence for hypertension and related conditions.
19. OBDRV19: Frequency of visits indicates disease management and severity.
20. PERWT19F: Obesity is a significant risk factor for hypertension; BMI can be derived from this.
21. TOTTCH19: Reflects the economic impact of hypertension and its comorbidities.

```python
# Specify column types based on your description
numerical_columns = ['HIBPAGED', 'EMPST31', 'FAMINC19', 'RXNUM', 'PERWT19F', 'TOTTCH19', 'OBDRV19']

categorical_columns = ['SEX', 'RACETHX', 'HIBPDX', 'BPMLDX', 'CHDDX', 'CHOLDX', 'CHOLAGED', 'STRKDX', 
 'MNHLTH31', 'ACTLIM31', 'DIABDX_M18', 'MIDX', 'OFTSMK53', 'INSCOV19']


# Convert data types
hyper_tension_df2[numerical_columns] = hyper_tension_df2[numerical_columns].apply(pd.to_numeric, errors='coerce')
hyper_tension_df2[categorical_columns] = hyper_tension_df2[categorical_columns].astype('float').astype('Int64').astype('category')

# Impute missing values: mean for numerical, mode for categorical
for col in numerical_columns:
    hyper_tension_df2[col] = hyper_tension_df2[col].fillna(hyper_tension_df2[col].mean())

for col in categorical_columns:
    hyper_tension_df2[col] = hyper_tension_df2[col].fillna(hyper_tension_df2[col].mode()[0])
```

```python
# Detect outliers
def detect_outliers(data):
    # Calculate Q1 (25th percentile), Q3 (75th percentile), and IQR for each column
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    # Calculate lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Return boolean DataFrame where True indicates an outlier
    return ((data < lower_bound) | (data > upper_bound))

# Select numerical columns only
num_cols = hyper_tension_df2.drop(['DUPERSID','SEX', 'RACETHX', 'HIBPDX', 'BPMLDX', 'CHDDX', 'CHOLDX', 'CHOLAGED', 'STRKDX', 
 'MNHLTH31', 'ACTLIM31', 'DIABDX_M18', 'MIDX', 'OFTSMK53', 'INSCOV19'],axis=1)

# Detect outliers in the numeric columns
outliers = detect_outliers(num_cols)

# Count the number of outliers for each row
outlier_count = outliers.sum(axis=1)

# Remove rows where more than 3 attributes have outlier values
hyper_tension_df3 = hyper_tension_df2[outlier_count <= 3]

# Number of rows left after removing outliers
print(f"Number of tuples before removing outliers: {hyper_tension_df2.shape[0]}")
remaining_rows = hyper_tension_df3.shape[0]
print(f"Number of tuples left after removing outliers: {remaining_rows}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import itertools

# List of categorical columns to test
columns = ['SEX', 'RACETHX', 'HIBPDX', 'BPMLDX', 'CHDDX', 'CHOLDX', 'CHOLAGED', 'STRKDX', 
 'MNHLTH31', 'ACTLIM31', 'DIABDX_M18', 'MIDX', 'OFTSMK53', 'INSCOV19']

# Generate all unique pairs of columns
pairs = list(itertools.combinations(columns, 2))

# Loop through each pair and perform the Chi-square test
for var1, var2 in pairs:
    contingency_table = pd.crosstab(hyper_tension_df3[var1], hyper_tension_df3[var2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f"Chi-square Test between {var1} and {var2}")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:")
    print(expected)
    print("\n" + "-"*40 + "\n")
```

OFTSMK53 has a high p-value so this is dropped and also CHOLAGED is highly correlated with CHOLDX. So,  CHOLAGED is dropped by keeping CHOLDX assuming it is directly associated with health outcomes.
